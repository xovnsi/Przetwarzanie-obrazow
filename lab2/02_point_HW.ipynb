# Zadanie domowe

W przypadku obrazów w odcieniach szarości pojedynczy piksel z zakresu [0; 255] reprezentowany jest jako 8-bitowa liczba bez znaku.
Pewnym rozszerzeniem analizy sposobu reprezentacji obrazu może być następujący eksperyment.
Załóżmy, że z każdego z 8 bitów możemy stworzyć pojedynczy obraz binarny (ang. _bit-plane slicing_).
Dla obrazka _100zloty.jpg_ (https://raw.githubusercontent.com/vision-agh/poc_sw/master/02_Point/100zloty.jpg) stwórz 8 obrazów, z których każdy powinien zawierać jedną płaszczyznę bitową.
Podpowiedź $-$ warto sprawdzić, jak realizuje się bitowe operacje logiczne.
Zastosowanie takiej dekompozycji obrazu pozwala na analizę ,,ważności'' poszczególnych bitów.
Jest to użyteczne w kwantyzacji, ale także w kompresji.

```python
import cv2
import os
import requests
import numpy as np
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/vision-agh/poc_sw/master/02_Point/'
fileName = '100zloty.jpg'
if not os.path.exists(fileName) :
    r = requests.get(url + fileName, allow_redirects=True)
    open(fileName, 'wb').write(r.content)
```

```python
pxl = []
pln = cv2.imread(fileName, 0)
for r in range(pln.shape[0]):
    for c in range(pln.shape[1]):
        pxl.append(np.binary_repr(pln[r][c], width=8)) #changing pixel value to binary representation
```

```python
plt.xticks([]), plt.yticks([])
plt.gray()
plt.imshow(pln)
```

```python
eight_bit = (np.array([int(p[0]) for p in pxl])).reshape(pln.shape[0],pln.shape[1])
#selecting the oldest bit
plt.xticks([]), plt.yticks([])
plt.gray()
plt.imshow(eight_bit)
```

```python
seventh_bit = (np.array([int(p[1]) for p in pxl])).reshape(pln.shape[0],pln.shape[1])
plt.xticks([]), plt.yticks([])
plt.gray()
plt.imshow(seventh_bit)
```

```python
sixth_bit = (np.array([int(p[2]) for p in pxl])).reshape(pln.shape[0],pln.shape[1])
plt.xticks([]), plt.yticks([])
plt.gray()
plt.imshow(sixth_bit)
```

```python
fifth_bit = (np.array([int(p[3]) for p in pxl])).reshape(pln.shape[0],pln.shape[1])
plt.xticks([]), plt.yticks([])
plt.gray()
plt.imshow(fifth_bit)
```

```python
fourth_bit = (np.array([int(p[4]) for p in pxl])).reshape(pln.shape[0],pln.shape[1])
plt.xticks([]), plt.yticks([])
plt.gray()
plt.imshow(fourth_bit)
```

```python
third_bit = (np.array([int(p[5]) for p in pxl])).reshape(pln.shape[0],pln.shape[1])
plt.xticks([]), plt.yticks([])
plt.gray()
plt.imshow(third_bit)
```

```python
second_bit = (np.array([int(p[6]) for p in pxl])).reshape(pln.shape[0],pln.shape[1])
plt.xticks([]), plt.yticks([])
plt.gray()
plt.imshow(second_bit)
```

```python
first_bit = (np.array([int(p[7]) for p in pxl])).reshape(pln.shape[0],pln.shape[1])
plt.xticks([]), plt.yticks([])
plt.gray()
plt.imshow(first_bit)
```

W drugim etapie zadania proszę spróbować odtworzyć obraz oryginalny z mniejszej liczby obrazów binarnych.
Warto zacząć od dwóch najbardziej znaczących bitów, a później dodawać kolejne.
Należy utworzyć co najmniej trzy wersje zrekonstruowanych obrazów.
Podpowiedź $-$ rekonstrukcja obrazu to mnożenie przez odpowiednią potęgę liczby 2 (przesunięcie bitowe) oraz dodawanie.

```python
rec = eight_bit*128+seventh_bit*64
plt.imshow(rec, cmap='gray')
```

```python
rec = fifth_bit*16+third_bit*4+second_bit*2
plt.imshow(rec, cmap='gray')
```

```python
rec = sixth_bit*32+fifth_bit*16
plt.imshow(rec, cmap='gray')
```
